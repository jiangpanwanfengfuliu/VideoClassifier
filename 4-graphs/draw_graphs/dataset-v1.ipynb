```python
import umap
import numpy as np
import pandas as pd
import torch
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.preprocessing import LabelEncoder
import os

# 加载数据集
X = torch.load(os.path.join(os.getcwd(), "train-data.pth"))
print(X.shape)
X = X.reshape(X.shape[0], -1)
Y = pd.read_csv(os.path.join(os.getcwd(), "train-labels.csv")
                ).class_label.to_list()
lb = LabelEncoder()
Y = lb.fit_transform(Y)
# 创建 UMAP 对象并设置参数
umap_model = umap.UMAP(n_neighbors=X.shape[0], n_components=2, random_state=42)

# 拟合并转换数据
X_umap = umap_model.fit_transform(X)
print(X_umap.shape)
```

```python
# 定义颜色映射
# colors = ['r', 'g', 'b']
labels_enc = np.array(list(set(Y)), dtype=int)
labels = lb.inverse_transform(labels_enc)
Y = np.array(Y, dtype=int)
# 创建散点图并添加图例
plt.figure(figsize=(8, 6))
for i in labels_enc:
    plt.scatter(X_umap[Y == i, 0],
                X_umap[Y == i, 1],  label=labels[i], s=50)
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.title('Dataset-v1-Raw')
plt.xlabel('UMAP Dimension 1')
plt.ylabel('UMAP Dimension 2')
plt.show()
```
